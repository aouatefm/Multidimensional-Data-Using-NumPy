```python
#Deep Copy
#The copy() method makes a complete copy of the array and its data, and doesnâ€™t share with the original array.
import numpy as np
```

```python
fruits = np.array(["Apple","Mango","Grapes","Watermelon"])
```

```python
#We now Create a deep copy of fruits 
```

```python
basket = fruits.copy()
```

```python
basket
```

```python
basket is fruits
```

```python
basket.base is fruits  # basket doesn't share anything with fruits
```

```python
#Change contents or shape of bakset. It does not change the contents of fruits
basket [0] = "Strawberry"
```

```python
basket
```

```python
fruits
```

```python
basket.shape = 2,2
```

```python
print("Shape of basket: ")
print(basket)
```
